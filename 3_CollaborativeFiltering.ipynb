```python
ratings = pd.read_csv('https://raw.githubusercontent.com/sidooms/MovieTweetings/master/latest/ratings.dat',
                names=['user_id','movie_id',
                        'rating','rating_timestamp'],
                 sep='::')
```

```python
ratings['user_id'] = ratings['user_id'].astype(str)
ratings['movie_id'] = ratings['movie_id'].astype(str)
```

```python
#get rid of rows that are NaN or for users who only have one rating
ratings = ratings.dropna()
ratings.describe()
```

```python
ratings = ratings[~ratings['user_id'].isin(ratings.groupby('user_id').count().query("movie_id == 1").index.tolist())]
```

```python
ratings['rating'] = ratings['rating'].astype(float)
```

```python
scaler = MinMaxScaler()
ratings[['rating_norm']] = scaler.fit_transform(ratings[['rating']])
```

```python
ratings.head()
```

```python
ratings['user_id'] = ratings['user_id'].astype('category')
ratings['movie_id'] = ratings['movie_id'].astype('category')
```

```python
coo = coo_matrix((ratings['rating_norm'].astype(float),
                  (ratings['movie_id'].cat.codes.copy(),
                   ratings['user_id'].cat.codes.copy())))
```

```python
csr = coo.tocsr()
```

```python
movies = dict(enumerate(ratings['movie_id'].cat.categories))
users = dict(enumerate(ratings['user_id'].cat.categories))
```

```python
users2inx = {v:k for k,v in users.items()}
movie2inx = {v:k for k,v in movies.items()}
```

```python
#factorization
```

```python
from sklearn.decomposition import NMF
```

```python
model = NMF(n_components=100, init='nndsvda', random_state=0, verbose=True)
```

```python
W = model.fit_transform(csr)
```

```python
H = model.components_.T
```

```python
def predict_recs(user_inx, take = 10):
    user_vec = H[user_inx]
    predictions = np.dot(user_vec, W.T)
    top = np.argsort(predictions)[::-1][:(take * 2)]
    return {movies.get(r, "0"): predictions[r] for r in top}

def pred_rating(user_id, movie_id):
    user_vec = H[users2inx[user_id]]
    movie_vec = W[movie2inx[movie_id]]
    return np.dot(user_vec, movie_vec)

predict_recs(users2inx['10097'], 10)
```

```python
def lookup_preds(row, user_inx=3):
    result = []
    for r in row:
        user_vec = H[user_inx]
        movie_vec = W[movie2inx[r]]
        result.append(np.dot(user_vec, movie_vec))
    return result
```

```python
def compare_ratings_predictions(user_id):
    
    user_x = ratings.loc[(ratings['user_id'] == user_id)]

    user_x.loc[:, "predictions"] = lookup_preds(user_x['movie_id'], users2inx[user_id])
    return user_x.sort_values('predictions', ascending=False)

compare_ratings_predictions('10097')
```
