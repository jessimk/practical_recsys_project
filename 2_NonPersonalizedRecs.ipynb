```python
fresh_ratingstop10 = fresh_ratings[['movie_id', 'rating']].groupby(
    'movie_id').count().sort_values(
    by='rating', ascending=False).reset_index().head(10)
fresh_ratingstop10.columns = ['movie_id', 'number_ratings']
fresh_ratingstop10
```

Top 10 Per Genre  

dictionary where key is genre : value is list of movie ids  

use this to filter ratings genre by list of movie ids

then use timestamp to get a relevant list of top 10 per genre

```python
unique_genres = list()

for ix in np.arange(0,movies['genre'].shape[0]-1,1):
    if type(movies['genre'][ix]) != list:
        pass
    else:
        unique_genres= list(set(list(unique_genres + list(set(movies['genre'][ix])))))
        
unique_genres
```

```python
movies.head()
```

```python
#create a dictionary of movie ids per genre

#initialize an empty dictionary
movies_by_genre = {}

#fill the dictionary with a movie id list per genre key
for genre in unique_genres:
    #create genre key_value pairing
    movies_by_genre[genre] = []
    
    for movie_ix in np.arange(0,movies.shape[0]-1,1):
        if type(movies['genre'][movie_ix]) != list:
            pass
        elif genre in movies['genre'][movie_ix]:
            movies_by_genre[genre].append(movies['movie_id'][movie_ix])
```

```python
    ratings[
    
    ratings.rating_dt.between(
    date.today() - relativedelta(months=1), date.today()) & 
        
        ratings['movie_id'].isin(movies_by_genre['Documentary'])][['movie_id','rating']].groupby(
    
    'movie_id').count().sort_values(by='rating', ascending=False).reset_index().head(10)['rating'].sum()
```

```python
horror_top10 = ratings[
    
    ratings.rating_dt.between(
    date.today() - relativedelta(months=1), date.today()) & 
        
        ratings['movie_id'].isin(movies_by_genre['Horror'])][['movie_id','rating']].groupby(
    
    'movie_id').count().sort_values(by='rating', ascending=False).reset_index().head(10)

horror_top10
```

```python
#create a dictionary of genres and a top10 df per genre

#initialize an empty dictionary

toptencharts = {}

for genre in unique_genres:
    
    total_fresh_ratings =     ratings[
    
        ratings.rating_dt.between(
        date.today() - relativedelta(months=1), date.today()) & 

            ratings['movie_id'].isin(movies_by_genre[genre])][['movie_id','rating']].groupby(

        'movie_id').count().sort_values(by='rating', ascending=False).reset_index().head(10)['rating'].sum()
    
    if total_fresh_ratings < 100:
        
        temp_df = ratings[ratings.rating_dt.between(
            date.today() - relativedelta(months=6), date.today()) & 

            ratings['movie_id'].isin(movies_by_genre[genre])][['movie_id','rating']].groupby(

            'movie_id').count().sort_values(by='rating', ascending=False).reset_index().head(10)
        
        temp_df = pd.merge(
                    temp_df, movies[['movie_id', 'movie_title']], 
                    how='left', on='movie_id')

        
        toptencharts[genre] = temp_df
        
    else:
        
        temp_df = ratings[ratings.rating_dt.between(
            date.today() - relativedelta(months=1), date.today()) & 

            ratings['movie_id'].isin(movies_by_genre[genre])][['movie_id','rating']].groupby(

            'movie_id').count().sort_values(by='rating', ascending=False).reset_index().head(10)
        
        temp_df = pd.merge(
                    temp_df, movies[['movie_id', 'movie_title']], 
                    how='left', on='movie_id')

        
        toptencharts[genre] = temp_df
```
